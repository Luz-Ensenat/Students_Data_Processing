# Student Grade Data Processing

We will use Pandas to manage and transform the data.

```python
import pandas as pd
```

In this project, we will work with a dataset containing information about students and their grades. The file *students-mat.csv* was obtained from the UCI Machine Learning Repository. To access the full dataset description, visit the following link: [Student Performance Data Set](https://archive.ics.uci.edu/ml/datasets/Student+Performance).

### 1. Data Loading

We read the CSV file called *students-mat.csv* and store it in a Pandas DataFrame named `students`.


```python
# We read the students file and store it in a DataFrame, specifying that the field separator is ';':

students = pd.read_csv("https://raw.githubusercontent.com/Luz-Ensenat/Students_Data_Processing/main/data/student-mat.csv", sep=';')

students
```

```python
# We inspect the number of rows and columns and print the result:

print("Rows and columns in the students DataFrame:", students.shape)
```

### 2. Filtering Underage Students

In this analysis, we will only consider students younger than 18 years old. Therefore, we overwrite the `students` DataFrame so that it only contains underage students.


```python
# We apply the filter to keep only underage students:

students = students[students['age'] < 18]

students
```

### 3. Renaming Columns

The columns G1, G2, and G3 contain the students' grades for three evaluation periods. We rename these columns to `grade1`, `grade2`, and `grade3`, respectively.


```python
# We rename the columns G1, G2, and G3 using the rename method:

students = students.rename(columns={'G1':'grade1','G2':'grade2','G3':'grade3'})

students
```

### 4. Column Filtering

We reduce the size of the DataFrame to keep only the columns relevant to our analysis. We retain the following columns: `school`, `sex`, `age`, `activities`, and the renamed columns `grade1`, `grade2`, and `grade3`.

We overwrite the `students` DataFrame with the filtered result.


```python
# We create a new DataFrame containing only the columns relevant to the analysis and overwrite the students DataFrame:

cols_keep = ['school','sex','age','activities','grade1','grade2','grade3']
students = students[cols_keep]

students
```

### 5. Calculation of Students Passing Each Evaluation Period

First, we create three boolean columns (`grade1_passed`, `grade2_passed`, `grade3_passed`) indicating whether the student passed each evaluation period.

A student passes if the grade is greater than or equal to 10.


```python
# We identify students who passed each grade by creating the three corresponding columns.
# We use loc to avoid the SettingWithCopyWarning error:

for i in (1, 2, 3):
    students.loc[:, f'grade{i}_passed'] = students[f'grade{i}'] >= 10

students
```

We can now count the number of students who passed each evaluation period.

We will determine how many students passed all three evaluation periods and how many failed all of them.


```python
# We calculate how many students passed all three evaluation periods and how many failed all of them:

passed_all = students[students[['grade1_passed','grade2_passed','grade3_passed']].all(axis=1)].shape[0]
failed_all = students[~students[['grade1_passed','grade2_passed','grade3_passed']].any(axis=1)].shape[0]

# We print the results:

print("Students who passed all three evaluation periods:", passed_all)
print("Students who failed all evaluation periods:", failed_all)
```

### 6. Average Grade per Student

We calculate the average grade for each student and add a new column to the DataFrame called `final_grade`.


```python
# We calculate the average grade for each student using mean and axis=1 so that the average is computed row by row.
# Once again, we use loc to avoid the SettingWithCopyWarning error:

students.loc[:, 'final_grade'] = students[['grade1', 'grade2', 'grade3']].mean(axis=1)

students
```

### 7. Average Final Grade by Age and Gender

We calculate the average final grade of students grouped by age and gender, rounding the result to two decimal places.

The result is stored in a DataFrame called `final_grade_stats`, and the output is sorted from highest to lowest average grade.


```python
# We group the data by age and gender, round the average grade values, and sort the results in descending order:

final_grade_stats = (
    students.groupby(['age','sex'], as_index=False)['final_grade']
    .mean()
    .round(2)
    .sort_values('final_grade', ascending=False)
)

print(final_grade_stats)
```

### 8. Duplicate Values

The `students` DataFrame may contain duplicate rows, so we identify and remove them.


```python
# We calculate the number of duplicate rows, print the result, and remove duplicates using drop_duplicates if necessary:

n_dup = students.duplicated().sum()

print("Duplicate rows:", n_dup)

if n_dup > 0:
    students = students.drop_duplicates()
    print("Duplicates removed.")
```
