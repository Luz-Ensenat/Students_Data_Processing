# Tratamiento de Datos de Notas de Estudiantes

Utilizaremos Pandas para manejar y transformar los datos

```python
import pandas as pd
```

En este proyecto vamos a trabajar con un conjunto de datos que contiene información sobre estudiantes y sus notas. El fichero *students-mat.csv* ha sido extraído del repositorio de Machine Learning de la UCI. Para consultar la descripción completa del conjunto de datos, se puede visitar el siguiente enlace: [Student Performance Data Set](https://archive.ics.uci.edu/ml/datasets/Student+Performance).

### 1. Carga de Datos

Leemos el fichero CSV llamado *students-mat.csv* y lo almacenamos en un DataFrame de Pandas llamado `students`.

```python
# Leemos el ficheros de students almacenándolo en un DataFrame, especificando que el separador de campos es ';':

students = pd.read_csv("https://raw.githubusercontent.com/Luz-Ensenat/Students_Data_Processing/main/data/student-mat.csv", sep=';')

students
```

```python
# Vemos el nº de filas y columnas y lo imprimimos:

print("Filas y columnas del DataFrame students:", students.shape)
```

### 2. Filtro estudiantes mayores de edad

En nuestro estudio vamos a considerar solo a los estudiantes menores de 18 años, por lo que sobrescribiremos el DataFrame `students` para que contendrá solo a los estudiantes menores de edad.

```python
# Aplicamos el fitro para dejar sólo estudiantes menores de edad:

students = students[students['age'] < 18]

students
```

### 3. Renombrar columnas

Las columnas G1, G2 y G3 contienen las notas de los estudiantes en tres cursos. Renombraremos estas columnas a `grade1`, `grade2` y `grade3` respectivamente.

```python
# Renombramos las columnas G1, G2 y G3 mediante el método rename:

students = students.rename(columns={'G1':'grade1','G2':'grade2','G3':'grade3'})

students
```

### 4. Filtrado de columnas

Vamos a reducir el tamaño del dataframe a solo las columnas que nos interesan. Mantendremos las siguientes columnas: `school`, `sex`, `age`, `activities` y las renombradas `grade1`, `grade2`, `grade3`. Sobrescribiremos el DataFrame `students` con el resultado.

```python
# Creamos un nuevo DataFrame sólo con las columnas que nos interesan y sobreescribimos students con él:

cols_keep = ['school','sex','age','activities','grade1','grade2','grade3']
students = students[cols_keep]

students
```

### 5. Calculo de aprobados por curso

Primero crearemos tres columnas de tipo booleano (`grade1_passed`, `grade2_passed`, `grade3_passed`) que indiquen si el estudiante ha aprobado cada uno de los cursos. Un estudiante aprueba si su nota es mayor o igual a 10.

```python
# Vemos los alumnos que han aprobado cada grade, creando las 3 columnas correspondientes y usamos loc para evitar el error SettingWithCopyWarning:

for i in (1, 2, 3):
    students.loc[:, f'grade{i}_passed'] = students[f'grade{i}'] >= 10

students
```

Ahora podemos contar el número de estudiantes que han aprobado cada curso. Veremos cuántos estudiantes han aprobado los tres cursos y cuántos no han aprobado ninguno

```python
# Medimos los que han aprobado los 3 cursos y los que no han aprobado ninguno:

passed_all   = students[students[['grade1_passed','grade2_passed','grade3_passed']].all(axis=1)].shape[0]
failed_all   = students[~students[['grade1_passed','grade2_passed','grade3_passed']].any(axis=1)].shape[0]

# Imprimimos los resultados:

print("Aprobados los 3 cursos:", passed_all)
print("No aprobó ninguno:",    failed_all)
```

### 6. Nota media por estudiante

Calcularemos la nota media de cada estudiante y añadiremos una nueva columna al DataFrame llamada `final_grade`.

```python
# Calculamos la nota media de cada alumno mediante mean y usando axis=1 para que calcule la media por fila. De nuevo usamos loc para evitar el error SettingWithCopyWarning:

students.loc[:, 'final_grade'] = students[['grade1', 'grade2', 'grade3']].mean(axis=1)

students
```

### 7. Nota media final por edad y sexo

Calcularemos la nota media final de los estudiantes por edad y sexo y redondearemos el resultado a dos decimales. Almacenaremos el resultado en un DataFrame llamado `final_grade_stats` y ordenamos el resultado por nota media de mayor a menor.

```python
# Agrupamos por edad y sexo, redondeamos la nota media y organizamos los resultados en order descendente:

final_grade_stats = (students.groupby(['age','sex'], as_index=False)['final_grade'].mean().round(2).sort_values('final_grade', ascending=False))
print(final_grade_stats)
```

### 8. Valores duplicados

Hay alguna fila duplicada en el DataFrame `students`, por lo que las eliminaremos.

```python
# Calculamos la cantidad de duplicados que hay, la imprimimos y, si es mayor que 0, eliminamos estos duplicados mediante drop:

n_dup = students.duplicated().sum()
print("Filas duplicadas:", n_dup)
if n_dup > 0:
    students = students.drop_duplicates()
    print("Duplicados eliminados.")
```
